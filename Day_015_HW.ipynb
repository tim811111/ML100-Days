```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
import random
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

random.seed(a=100)
ran_num100 = np.array([random.uniform(-1,1) for i in range(100)])
ran_np_10_10 = np.reshape(ran_num100,(10,10))

ran_num3000 = np.array([random.uniform(-1,1) for i in range(3000)])
ran_np_3_1000 = np.reshape(ran_num3000,(1000,3))

ran_n = np.random.randn(3000)
ran_n_3_1000 = np.reshape(ran_n,(1000,3))

print(ran_np_10_10.shape)
print(ran_np_3_1000.shape)
print(ran_n_3_1000.shape)
```

```python
"""
Your Code Here
"""
matrix = ran_np_10_10

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix, annot = True, square =True,cmap = plt.cm.RdYlBu_r)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = ran_np_3_1000

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(sns.scatterplot , alpha = 0.1)
grid.map_diag(sns.kdeplot, color = "r" )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r, legend = True, )

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = ran_n_3_1000

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(sns.scatterplot , alpha = 0.1)
grid.map_diag(sns.kdeplot, color = "r" )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r, legend = True, )

plt.show()
```
