### 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
from sklearn.cluster import AgglomerativeClustering
estimator = [('iris_ward', AgglomerativeClustering(n_clusters=4,linkage = 'ward')),
             ('iris_complete', AgglomerativeClustering(n_clusters=4,linkage='complete'))]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig_num = 1
titles = ['ward','complete']
for name, est in estimator:
    fig = plt.figure(fig_num, figsize= (16,9))
    ax = Axes3D(fig, rect=[0,0,0.95,1],elev=48, azim = 134)
    est.fit(X)
    lable = est.labels_
    ax.scatter(X[:,0], X[:,1],X[:,2],c=lable.astype(np.float),edgecolor='k')
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fig_num - 1])
    ax.dist = 12
    fig_num = fig_num + 1


for name, label in [('cls0', 0),
                    ('cls1', 1),
                    ('cls2', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean(),
              X[y == label, 2].mean() + 2, name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))

y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:,0], X[:,1],X[:,2], c=y, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('Ground Truth')
ax.dist = 12

#fig.show()
```
